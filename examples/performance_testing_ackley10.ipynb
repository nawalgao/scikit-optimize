```python
import numpy as np
np.random.seed(123)
%matplotlib inline
import seaborn as sns
sns.set_context("talk", font_scale=1.2)
import matplotlib.pyplot as plt
from skopt import Optimizer
from skopt.benchmarks2 import ackley
from pyDOE import *
```

```python
from scipy.spatial import distance
def euclidean(p_pred, p_act):
    """
    Euclidean distance
    """
    dst = distance.euclidean(p_pred, p_act)
    return dst
```

```python
def neg_ackley(X_train):
    Y = ackley(X_train)
    return Y
```

```python
ntrain = 20
ndim = 10
iterations = 20
```

```python
X_train_norm = lhs(ndim, samples= ntrain, criterion='maximin') # Initial training points
X_train = np.zeros(X_train_norm.shape)
X_train = X_train_norm*45 - 15.
Y_train = ackley(X_train)
Y_train_neg = neg_ackley(X_train)
```

```python
bounds = [(-15.0, 30.0)]*ndim
optimum = np.zeros(ndim)
```

# Constant Liar Strategy

# 1. GP

```python
opt = Optimizer(bounds, base_estimator="gp", acq_func = 'EI',
                acq_optimizer = 'sampling', random_state = 123)
opt.tell(X_train.tolist(), Y_train.tolist())
```

```python
import timeit
Xx = X_train
Yy = Y_train
all_points = []
time_tk = []
best_iter = []
for i in range(iterations):
    start = timeit.default_timer()
    next_points = opt.ask(n_points= 10)
    stop = timeit.default_timer()
    t = stop - start
    time_tk.append(t)
    next_obj = ackley(np.array(next_points))
    cur_best = np.abs(np.min(next_obj))
    best_iter.append(cur_best)
    Xx = np.vstack([Xx, np.array(next_points)])
    Yy = np.append(Yy, next_obj)
    opt.tell(Xx.tolist(), Yy.tolist())
    print ('Time taken for queries' + ':' + str(t)) 
    all_points.append(next_points)
    
gp_best = best_iter
gp_time = time_tk
gp_points = all_points
```

# 2. RF

```python
opt = Optimizer(bounds, base_estimator="rf", acq_func = 'EI',
                acq_optimizer = 'sampling', random_state = 123)
opt.tell(X_train.tolist(), Y_train.tolist())

import timeit
Xx = X_train
Yy = Y_train
all_points = []
time_tk = []
best_iter = []
for i in range(iterations):
    start = timeit.default_timer()
    next_points = opt.ask(n_points= 10)
    stop = timeit.default_timer()
    t = stop - start
    time_tk.append(t)
    next_obj = ackley(np.array(next_points))
    cur_best = np.abs(np.min(next_obj))
    best_iter.append(cur_best)
    Xx = np.vstack([Xx, np.array(next_points)])
    Yy = np.append(Yy, next_obj)
    opt.tell(Xx.tolist(), Yy.tolist())
    print ('Time taken for queries' + ':' + str(t)) 
    all_points.append(next_points)
    
rf_best = best_iter
rf_time = time_tk
rf_points = all_points
```
