```python
import numpy as np
np.random.seed(123)
%matplotlib inline
import seaborn as sns
sns.set_context("talk", font_scale=1.2)
import matplotlib.pyplot as plt
from skopt import Optimizer
from skopt.benchmarks2 import ackley
from pyDOE import *
```

```python
def neg_ackley(X_train):
    Y = ackley(X_train)
    return Y
```

```python
ntrain = 200
ndim = 30
```

```python
X_train_norm = lhs(ndim, samples= ntrain, criterion='maximin') # Initial training points
X_train = np.zeros(X_train_norm.shape)
X_train = X_train_norm*45 - 15.
Y_train = ackley(X_train)
Y_train_neg = neg_ackley(X_train)
```

```python
bounds = [(-15.0, 30.0)]*ndim
```

# Constant Liar Strategy Save

# RF

```python
num_queries = [10,30,100, 300, 500, 1000]
```

```python
opt = Optimizer(bounds, base_estimator="rf", acq_func = 'EI',
                acq_optimizer = 'sampling', random_state = 123, )
opt.tell(X_train.tolist(), Y_train.tolist())
```

```python
import timeit

all_points = []
time_tk = []
for query in num_queries:
    start = timeit.default_timer()
    next_points = opt.ask(n_points= query)
    stop = timeit.default_timer()
    t = stop - start
    time_tk.append(t)
    print ('Time taken for queries' + str(query) + ':' + str(t)) 
    all_points.append(next_points)

np.savetxt('cl_rf_ack30.txt', time_tk)
```

```python
cl_rf_time = time_tk
```

```python
opt = Optimizer(bounds, base_estimator="rf", acq_func = 'qEI',
                acq_optimizer = 'sampling', random_state = 123)
opt.tell(X_train.tolist(), Y_train.tolist())

import timeit

all_points = []
time_tk = []
for query in num_queries:
    qei_args_dict = {'num_sampled_points': query,
               'num_batches_eval' : 4000,
               'strategy_batch_selection' : 'random'}
    start = timeit.default_timer()
    next_points = opt.ask(additional_acq_func_kwargs = qei_args_dict)
    stop = timeit.default_timer()
    t = stop - start
    time_tk.append(t)
    print ('Time taken for queries' + str(query) + ':' + str(t)) 
    all_points.append(next_points)

np.savetxt('qEI_N4000_rf_ack30.txt', time_tk)
```

```python
qEI_4000_time_tk = time_tk
```

```python
opt = Optimizer(bounds, base_estimator="rf", acq_func = 'qEI',
                acq_optimizer = 'sampling', random_state = 123)
opt.tell(X_train.tolist(), Y_train.tolist())

import timeit

all_points = []
time_tk = []
for query in num_queries:
    qei_args_dict = {'num_sampled_points': query,
               'num_batches_eval' : 400,
               'strategy_batch_selection' : 'random'}
    start = timeit.default_timer()
    next_points = opt.ask(additional_acq_func_kwargs = qei_args_dict)
    stop = timeit.default_timer()
    t = stop - start
    time_tk.append(t)
    print ('Time taken for queries' + str(query) + ':' + str(t)) 
    all_points.append(next_points)

np.savetxt('qEI_N400_rf_ack30.txt', time_tk)
```

```python
qEI_400_time_tk = time_tk
```

```python
x = np.array([10,30,100, 300, 500, 1000])
plt.figure(figsize = (15,12))
plt.plot(x, cl_rf_time, label = 'CL rf')
plt.plot(x, qEI_4000_time_tk, label = 'qEI rf 4000')
plt.plot(x, qEI_400_time_tk, label = 'qEI rf 300')
plt.xlabel('Number of next selected points')
plt.ylabel('Time in secs')
plt.legend(loc = 'best')
plt.title('Scalability : Ackley : 30 dimensions')
```
