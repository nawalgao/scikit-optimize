```python
import numpy as np
np.random.seed(123)
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context("talk", font_scale=1.2)
%matplotlib inline
```

```python
# Const Liar Branin
cl_gp_branin = np.loadtxt('cl_gp_branin.txt')
cl_rf_branin = np.loadtxt('cl_rf_branin.txt')
cl_llbnn_branin_300 = np.loadtxt('cl_llbnn_branin_300.txt')
cl_llbnn_branin_500 = np.loadtxt('cl_llbnn_branin_500.txt')

# Const Liar Hartmann
cl_gp_hart = np.loadtxt('cl_gp_hart.txt')
cl_rf_hart = np.loadtxt('cl_rf_hart.txt')
cl_llbnn_hart_300 = np.loadtxt('cl_llbnn_hart_300.txt')
cl_llbnn_hart_500 = np.loadtxt('cl_llbnn_hart_500.txt')

# qEI Branin
qEI_gp_branin_400 = np.loadtxt('qEI_N400_gp_branin.txt')
qEI_rf_branin_400 = np.loadtxt('qEI_N400_rf_branin.txt')
qEI_gp_branin_4000 = np.loadtxt('qEI_N4000_gp_branin.txt')
qEI_rf_branin_4000 = np.loadtxt('qEI_N4000_rf_branin.txt')

# qEI Hartmann

qEI_gp_hart_400 = np.loadtxt('qEI_N400_gp_hart.txt')
qEI_rf_hart_400 = np.loadtxt('qEI_N400_rf_hart.txt')
qEI_gp_hart_4000 = np.loadtxt('qEI_N4000_gp_hart.txt')
qEI_rf_hart_4000 = np.loadtxt('qEI_N4000_rf_hart.txt')
```

```python
x = np.array([10,30,100, 300])
plt.figure(figsize = (15,12))
plt.plot(x, cl_gp_branin, label = 'cl gp branin')
plt.plot(x, cl_rf_branin,'--', label = 'cl rf branin')
plt.plot(x, cl_llbnn_branin_300, label = 'cl llbnn 300 branin')
plt.plot(x, cl_llbnn_branin_500, label = 'cl llbnn 500 branin')
plt.plot(x, cl_gp_hart, label = 'cl gp hart')
plt.plot(x, cl_rf_hart, label = 'cl rf hart')
plt.plot(x, cl_llbnn_hart_300, label = 'cl llbnn 300 hart')
plt.plot(x, cl_llbnn_hart_500, label = 'cl llbnn 500 hart')

plt.plot(x, qEI_gp_branin_400[:4], label = 'qEI gp branin 400')
plt.plot(x, qEI_rf_branin_400[:4], label = 'qEI rf branin 400')
plt.plot(x, qEI_gp_branin_4000[:4], label = 'qEI gp branin 4000')
plt.plot(x, qEI_rf_branin_4000[:4], label = 'qEI rf branin 4000')


plt.plot(x, qEI_gp_hart_400[:4], label = 'qEI gp hart 400')
plt.plot(x, qEI_rf_hart_400[:4], label = 'qEI rf hart 400')
plt.plot(x, qEI_gp_hart_4000[:4], label = 'qEI gp hart 4000')
plt.plot(x, qEI_rf_hart_4000[:4], label = 'qEI rf hart 4000')
plt.xlabel('Number of next selected points')
plt.ylabel('Time in secs')
plt.legend(loc = 'best')
```
